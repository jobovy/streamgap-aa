```python
def load_denis_file(filename):
    names = ['x','y','z','vx','vy','vz','id']
    data = pd.read_csv(filename,names=names,comment='#',sep=' ')
    data['R']=np.sqrt(data['x']**2+data['y']**2)
    data['phi']=np.arctan2(data['y'],data['x'])
    data['vR']=(data['x']*data['vx']+data['y']*data['vy'])/data['R']
    data['vphi']=(-data['y']*data['vx']+data['x']*data['vy'])/data['R']
    data['E']=map(lambda i:pot.energy(i[:6]),data.values)
    data['Lz']=np.fabs(data['vphi']*data['R'])
    return data
```

```python
def _rotate_to_arbitrary_vector(a,b):
    v = np.cross(a,b)
    s = np.linalg.norm(v)
    c = np.dot(a,b)
    vx = np.zeros((3,3))
    vx[1][0],vx[0][1]=v[2],-v[2]
    vx[2][0],vx[0][2]=-v[1],v[1]
    vx[1][2],vx[2][1]=-v[0],v[0]
    return np.eye(3,3)+vx+np.dot(vx,vx)*(1-c)/s/s
from scipy.optimize import leastsq
def fitPlaneOptimize(X,guess=None):
    def residuals(parameter):
        n = parameter/np.sqrt(np.sum(parameter**2))
        return [np.dot(n,i) for i in X]
    p0=0
    if(isinstance(guess,np.ndarray)):
        p0=guess
    else:
        p0 = np.cross(X[0],X[-1]) # initial guess
    result = leastsq(residuals, p0)[0]
    return result/np.sqrt(np.sum(result**2))
def rotate_data(data,centre,guess=None):
    pp = fitPlaneOptimize(data,guess)
    LRot = _rotate_to_arbitrary_vector(pp,[0.,0.,1.])
    return np.array([map(lambda i:np.dot(LRot,i),data)]),np.dot(LRot,centre),pp
def rotate_data_pp(data,centre,pp):
    LRot = _rotate_to_arbitrary_vector(pp,[0.,0.,1.])
    return np.array([map(lambda i:np.dot(LRot,i),data)]),np.dot(LRot,centre),pp
```

```python
def measure_angles(v1,v2,v3):
    # Calculate the mean of the points, i.e. the 'center' of the cloud
    aa = np.vstack((v1,v2,v3)).T
    datamean = aa.mean(axis=0)
    # Do an SVD on the mean-centered data.
    uu, dd, vv = np.linalg.svd(aa - datamean,full_matrices=False)
    vv = vv.conj().transpose()
    print vv
    phi = np.arcsin(vv.T[0][2])
    theta = np.arccos(vv.T[0][1]/np.cos(phi))
    aa = aa - datamean
    plt.plot(aa.T[0],aa.T[1],'.',ms=4)
    plt.plot(aa.T[0],aa.T[2],'.',ms=4)
    plt.plot(np.linspace(np.min(aa.T[0]),np.max(aa.T[0])),-np.linspace(np.min(aa.T[0]),np.max(aa.T[0]))*1./np.tan(theta))
    plt.plot(np.linspace(np.min(aa.T[0]),np.max(aa.T[0])),-np.linspace(np.min(aa.T[0]),np.max(aa.T[0]))*np.tan(phi)/np.sin(theta))
    return phi,theta
```

## Construct model

Parameters chosen by hand

There are several modifications to the simple angle-frequency model. We use a non-uniform stripping rate that consists of two components. The first component is a steady evaporation component that has a narrower frequency distribution with a smaller offset from the progenitor and the second component is a comb of Gaussians centred on pericentric passage times. This second component has a broader and more offset frequency distribution. Both of these distributions are enveloped by a stripping rate distribution that increases linearly with stripping time.

```python
from scipy.interpolate import LSQUnivariateSpline, interp1d
from scipy.stats import linregress

class sub_halo:
    def __init__(self,M,b,rs,w,gc):
        G=4.300918e-6
        self.M=M
        self.GM = G*M
        self.b = b
        self.rs = rs
        self.w = w
        self.magw = np.sqrt(np.sum(w**2))
        self.gap_centre_impact = gc

class angle_freq_kicks:
            
    def dOmAngdv(self,x,dv):
        A=AA.angles(x)
        dOdv=np.zeros((3,3))
        dAdv=np.zeros((3,3))
        AOu=np.zeros((3,6))
        AOd=np.zeros((3,6))
        xu,xd=np.copy(x),np.copy(x)
        xu[3]+=dv
        xd[3]-=dv
        AOu[0]=AA.angles(xu)
        AOd[0]=AA.angles(xd)
        xu,xd=np.copy(x),np.copy(x)
        xu[4]+=dv
        xd[4]-=dv
        AOu[1]=AA.angles(xu)
        AOd[1]=AA.angles(xd)
        xu,xd=np.copy(x),np.copy(x)
        xu[5]+=dv
        xd[5]-=dv
        AOu[2]=AA.angles(xu)
        AOd[2]=AA.angles(xd)
        dAdv=AOu.T[:3].T-AOd.T[:3].T
        dOdv=AOu.T[3:6].T-AOd.T[3:6].T
        l,e = np.linalg.eigh(dAdv.T/2./dv)
        return A[:3],A[3:],dAdv.T/2./dv,dOdv.T/2./dv,e[-2]

        
    def plot_kicks(self):
        f,a=plt.subplots(1,3,figsize=(15,5));plt.subplots_adjust(wspace=0.4);
        a[0].plot(self.deltaang_par,self.dv.T[0],label=r'$\delta v^g_x$');
        a[0].plot(self.deltaang_par,self.dv.T[1],label=r'$\delta v^g_y$');
        a[0].plot(self.deltaang_par,self.dv.T[2],label=r'$\delta v^g_z$');
        a[0].set_xlabel(labels['dangpar']);a[0].set_ylabel(labels['dvg']);a[0].legend(loc=3)

        a[1].plot(self.deltaang_par,self.dA.T[0],label=r'$\delta \theta^g_R$');
        # a[1].plot(afk.deltaang_par,afk.angR_k(afk.deltaang_par));
        a[1].plot(self.deltaang_par,self.dA.T[1],label=r'$\delta \theta^g_\phi$');
        # a[1].plot(afk.deltaang_par,afk.angp_k(afk.deltaang_par));
        a[1].plot(self.deltaang_par,self.dA.T[2],label=r'$\delta \theta^g_z$')
        # a[1].plot(afk.deltaang_par,afk.angz_k(afk.deltaang_par));
        a[1].set_xlabel(labels['dangpar']);a[1].set_ylabel(labels['dangg']);a[1].legend(loc=3)

        a[2].plot(self.deltaang_par,self.dO.T[0],label=r'$\delta \Omega^g_R$');
        # a[2].plot(afk.deltaang_par,afk.OmR_k(afk.deltaang_par));
        a[2].plot(self.deltaang_par,self.dO.T[1],label=r'$\delta \Omega^g_\phi$');
        # a[2].plot(afk.deltaang_par,afk.Omp_k(afk.deltaang_par));
        a[2].plot(self.deltaang_par,self.dO.T[2],label=r'$\delta \Omega^g_z$')
        # a[2].plot(afk.deltaang_par,afk.Omz_k(afk.deltaang_par));
        a[2].set_xlabel(labels['dangpar']);a[2].set_ylabel(labels['dOmg']);a[2].legend(loc=2)
    
    
    def __init__(self,merged_peri,sub_halo_props,num=210):
        
        merged_peri['phi']=np.arctan2(merged_peri['y'],merged_peri['x'])
        merged_peri['ThR']=merged_peri['ThR']-2.*np.pi*(merged_peri['ThR']>np.pi)
        merged_peri['Thz']=merged_peri['Thz']-2.*np.pi*(merged_peri['Thz']>np.pi)
        merged_peri['phi_int']=merged_peri['phi']-2.*np.pi*(merged_peri['phi']>0.)
        
        ## Build spline of track
        f,a=plt.subplots(2,6,figsize=(20,7));plt.subplots_adjust(wspace=0.7,hspace=0.3)
        xs = np.linspace(np.min(merged_peri['phi_int']),np.max(merged_peri['phi_int']),num)
        knots = np.linspace(np.min(merged_peri['phi_int']),np.max(merged_peri['phi_int']),12)[1:-1]

        sx=LSQUnivariateSpline(merged_peri['phi_int'][np.argsort(merged_peri['phi_int'])].values,merged_peri['x'][np.argsort(merged_peri['phi_int'])].values,knots)
        hi=a[0][0].plot(np.sort(merged_peri['phi_int']),merged_peri['x'][np.argsort(merged_peri['phi_int'])],'.',ms=3)
        a[0][0].plot(xs,sx(xs));a[0][0].set_xlabel(labels['x']);a[0][0].set_ylabel(labels['x']);
        
        s=LSQUnivariateSpline(merged_peri['phi_int'][np.argsort(merged_peri['phi_int'])].values,merged_peri['y'][np.argsort(merged_peri['phi_int'])].values,knots)
        hi=a[0][1].plot(np.sort(merged_peri['phi_int']),merged_peri['y'][np.argsort(merged_peri['phi_int'])],'.',ms=3)
        a[0][1].plot(xs,s(xs));a[0][1].set_xlabel(labels['x']);a[0][1].set_ylabel(labels['y']);
             
        sz=LSQUnivariateSpline(merged_peri['phi_int'][np.argsort(merged_peri['phi_int'])].values,merged_peri['z'][np.argsort(merged_peri['phi_int'])].values,knots)
        hi=a[0][2].plot(np.sort(merged_peri['phi_int']),merged_peri['z'][np.argsort(merged_peri['phi_int'])],'.',ms=3)
        a[0][2].plot(xs,sz(xs));a[0][2].set_xlabel(labels['x']);a[0][2].set_ylabel(labels['z']);
        
        s1=LSQUnivariateSpline(merged_peri['phi_int'][np.argsort(merged_peri['phi_int'])].values,merged_peri['vx'][np.argsort(merged_peri['phi_int'])].values,knots)
        hi=a[0][3].plot(np.sort(merged_peri['phi_int']),merged_peri['vx'][np.argsort(merged_peri['phi_int'])],'.',ms=3)
        a[0][3].plot(xs,s1(xs));a[0][3].set_xlabel(labels['x']);a[0][3].set_ylabel(labels['vx']);
        
        s2=LSQUnivariateSpline(merged_peri['phi_int'][np.argsort(merged_peri['phi_int'])].values,merged_peri['vy'][np.argsort(merged_peri['phi_int'])].values,knots)
        hi=a[0][4].plot(np.sort(merged_peri['phi_int']),merged_peri['vy'][np.argsort(merged_peri['phi_int'])],'.',ms=3)
        a[0][4].plot(xs,s2(xs));a[0][4].set_xlabel(labels['x']);a[0][4].set_ylabel(labels['vy']);
             
        s2z=LSQUnivariateSpline(merged_peri['phi_int'][np.argsort(merged_peri['phi_int'])].values,merged_peri['vz'][np.argsort(merged_peri['phi_int'])].values,knots)
        hi=a[0][5].plot(np.sort(merged_peri['phi_int']),merged_peri['vz'][np.argsort(merged_peri['phi_int'])],'.',ms=3)
        a[0][5].plot(xs,s2z(xs));a[0][5].set_xlabel(labels['x']);a[0][5].set_ylabel(labels['vz']);
        
        stream_track = np.ones((len(xs),6))*1e-5
        for i,j in enumerate(xs):
            stream_track[i][0]=sx(j)
            stream_track[i][1]=s(j)
            stream_track[i][2]=sz(j)
            stream_track[i][3]=s1(j)
            stream_track[i][4]=s2(j)
            stream_track[i][5]=s2z(j)
        a[0][0].plot(xs,stream_track.T[0])
        a[0][1].plot(xs,stream_track.T[1])
        a[0][2].plot(xs,stream_track.T[2])
        a[0][3].plot(xs,stream_track.T[3])
        a[0][4].plot(xs,stream_track.T[4])
        a[0][5].plot(xs,stream_track.T[5])
        
        dOdAdv=map(lambda i:self.dOmAngdv(i,0.5),stream_track)
        self.dOdAdv = dOdAdv
        self.subhalo=sub_halo_props
        b = sub_halo_props.b
        w = sub_halo_props.w
        GM = sub_halo_props.GM
        rs = sub_halo_props.rs
        gci = sub_halo_props.gap_centre_impact
        self.dv = galpy.df_src.streamgapdf.impulse_deltav_plummer_general(stream_track.T[3:].T,stream_track.T[:3].T,b,w,gci[:3],gci[3:],GM,rs)
        self.dA = np.array([map(lambda i,j:np.dot(i[2],j),dOdAdv,self.dv)])[0]
        self.dO = np.array([map(lambda i,j:np.dot(i[3],j),dOdAdv,self.dv)])[0]
        self.O =  np.array([map(lambda i:i[1],dOdAdv)])[0]
        self.A = np.array([map(lambda i:i[0],dOdAdv)])[0]
        self.e = np.array([map(lambda i:i[4],dOdAdv)])[0]
        for i in [0,2]:
            self.A.T[i]=self.A.T[i]-2.*np.pi*(self.A.T[i]>np.pi)
        a[1][0].plot(self.A.T[0],self.A.T[1])
        a[1][1].plot(self.A.T[0],self.A.T[2])
        a[1][2].plot(self.O.T[0],self.O.T[1])
        a[1][3].plot(self.O.T[1],self.O.T[2])
        slope, intercept, r_value, p_value, std_err = linregress(self.O.T[0],self.O.T[1])
        slope2, intercept2, r_value, p_value, std_err = linregress(self.O.T[0],self.O.T[2])
        self.n = np.array([1.,slope,slope2])
        a[1][2].plot(self.O.T[0],self.n[1]*self.O.T[0]+intercept)
        a[1][3].plot(self.n[1]*self.O.T[0]+intercept,self.n[2]*self.O.T[0]+intercept2)
        self.n = self.n/np.sqrt(np.sum(self.n**2))
        self.acts_centre = AA.actions(gci)
        self.angs_centre = AA.angles(gci)
        for i in [0,2]:
            self.angs_centre[i]=self.angs_centre.T[i]-2.*np.pi*(self.angs_centre.T[i]>np.pi)
        self.ndotac = np.dot(self.n,self.angs_centre[:3])
        self.ndotfc = np.dot(self.n,self.angs_centre[3:])
        self.deltaang_par=np.dot(self.A-self.angs_centre[:3],self.n)
#         self.deltaang_par=self.deltaang_par-2.*np.pi*(self.deltaang_par>np.pi)
        self.vdotdv = np.array(map(lambda i:np.dot(i[0],i[1]),zip(stream_track.T[3:].T,self.dv)))
        self.stream_track=stream_track
        a[1][4].plot(self.A.T[0])
        self.angR_k=interp1d(self.deltaang_par[np.fabs(self.dA.T[0])<1.],self.dA.T[0][np.fabs(self.dA.T[0])<1.])
        self.angp_k=interp1d(self.deltaang_par[np.fabs(self.dA.T[1])<1.],self.dA.T[1][np.fabs(self.dA.T[1])<1.])
        self.angz_k=interp1d(self.deltaang_par[np.fabs(self.dA.T[2])<1.],self.dA.T[2][np.fabs(self.dA.T[2])<1.])
        self.OmR_k=interp1d(self.deltaang_par,self.dO.T[0])
        self.Omp_k=interp1d(self.deltaang_par,self.dO.T[1])
        self.Omz_k=interp1d(self.deltaang_par,self.dO.T[2])
        self.plot_kicks()
    
    def kick(self,ang):
        return self.angR_k(ang),self.angp_k(ang),self.angz_k(ang),self.OmR_k(ang),self.Omp_k(ang),self.Omz_k(ang)

    def kick_unperturbed(self,data2):
        data=data2.copy()
        data['dangpar']=np.dot(self.n,np.array([data['ThR'].T,data['Thp'].T,data['Thz'].T]))-self.ndotac
        mm = np.maximum(np.min(self.deltaang_par),np.minimum(np.max(self.deltaang_par),data['dangpar']))
        data['angR_k']=self.angR_k(mm)
        data['angp_k']=self.angp_k(mm)
        data['angz_k']=self.angz_k(mm)
        data['OmR_k']=self.OmR_k(mm)
        data['Omp_k']=self.Omp_k(mm)
        data['Omz_k']=self.Omz_k(mm)
        return data

    def new_kick(self,sub_halo_props):
        self.subhalo=sub_halo_props
        b = sub_halo_props.b
        w = sub_halo_props.w
        GM = sub_halo_props.GM
        rs = sub_halo_props.rs
        gci = sub_halo_props.gap_centre_impact
        self.dv = galpy.df_src.streamgapdf.impulse_deltav_plummer_general(self.stream_track.T[3:].T,self.stream_track.T[:3].T,b,w,gci[:3],gci[3:],GM,rs)
        self.dA = np.array([map(lambda i,j:np.dot(i[2],j),self.dOdAdv,self.dv)])[0]
        self.dO = np.array([map(lambda i,j:np.dot(i[3],j),self.dOdAdv,self.dv)])[0]
        self.angs_centre = AA.angles(gci)
        for i in [0,2]:
            self.angs_centre[i]=self.angs_centre.T[i]-2.*np.pi*(self.angs_centre.T[i]>np.pi)
        self.ndotac = np.dot(self.n,self.angs_centre[:3])
        self.ndotfc = np.dot(self.n,self.angs_centre[3:])
        self.deltaang_par=np.dot(self.A-self.angs_centre[:3],self.n)
        self.vdotdv = np.array(map(lambda i:np.dot(i[0],i[1]),zip(self.stream_track.T[3:].T,self.dv)))
        self.angR_k=interp1d(self.deltaang_par,self.dA.T[0])
        self.angp_k=interp1d(self.deltaang_par,self.dA.T[1])
        self.angz_k=interp1d(self.deltaang_par,self.dA.T[2])
        self.OmR_k=interp1d(self.deltaang_par,self.dO.T[0])
        self.Omp_k=interp1d(self.deltaang_par,self.dO.T[1])
        self.Omz_k=interp1d(self.deltaang_par,self.dO.T[2])
```

```python
def kick_and_evolve_forwards(data,kick,t):
    data2=kick.kick_unperturbed(data)
    data2['ThRk']=data2['ThR']+data2['angR_k']
    data2['Thpk']=data2['Thp']+data2['angp_k']
    data2['Thzk']=data2['Thz']+data2['angz_k']
    data2['OmRk']=data2['OmR']+data2['OmR_k']
    data2['Ompk']=data2['Omp']+data2['Omp_k']
    data2['Omzk']=data2['Omz']+data2['Omz_k']
    data2['x2']=data['x']
    data2['y2']=data['y']
    data2['z2']=data['vz']
    data2['vx2']=data['vx']
    data2['vy2']=data['vy']
    data2['vz2']=data['vz']
    data2['ThR2']=data['ThR']
    data2['Thp2']=data['Thp']
    data2['Thz2']=data['Thz']
    data2['DThR2']=data['DThR']
    data2['DThp2']=data['DThp']
    data2['DThz2']=data['DThz']
    data2['Thpar2']=data['Thpar']
    data2['Thperp2']=data['Thperp']
    data2['x2_u']=data['x']
    data2['y2_u']=data['y']
    data2['z2_u']=data['vz']
    data2['vx2_u']=data['vx']
    data2['vy2_u']=data['vy']
    data2['vz2_u']=data['vz']
    data2['ThR2_u']=data['ThR']
    data2['Thp2_u']=data['Thp']
    data2['Thz2_u']=data['Thz']
    data2['DThR2_u']=data['DThR']
    data2['DThp2_u']=data['DThp']
    data2['DThz2_u']=data['DThz']
    data2['Thpar2_u']=data['Thpar']
    data2['Thperp2_u']=data['Thperp']
    for i in range(len(data)):
        D = sampler.evolve(np.array([data2['ThRk'][i],data2['Thpk'][i],data2['Thzk'][i]]),
                           np.array([data2['OmRk'][i],data2['Ompk'][i],data2['Omzk'][i]]),
                           t)
        data2['x2'][i]=D[0]
        data2['y2'][i]=D[1]
        data2['z2'][i]=D[2]
        data2['vx2'][i]=D[3]
        data2['vy2'][i]=D[4]
        data2['vz2'][i]=D[5]
        data2['ThR2'][i]=D[9]
        data2['Thp2'][i]=D[10]
        data2['Thz2'][i]=D[11]
        data2['DThR2'][i]=D[15]
        data2['DThp2'][i]=D[16]
        data2['DThz2'][i]=D[17]
        data2['Thpar2'][i]=D[20]
        data2['Thperp2'][i]=D[21]
        D = sampler.evolve(np.array([data2['ThR'][i],data2['Thp'][i],data2['Thz'][i]]),
                           np.array([data2['OmR'][i],data2['Omp'][i],data2['Omz'][i]]),
                           t)
        data2['x2_u'][i]=D[0]
        data2['y2_u'][i]=D[1]
        data2['z2_u'][i]=D[2]
        data2['vx2_u'][i]=D[3]
        data2['vy2_u'][i]=D[4]
        data2['vz2_u'][i]=D[5]
        data2['ThR2_u'][i]=D[9]
        data2['Thp2_u'][i]=D[10]
        data2['Thz2_u'][i]=D[11]
        data2['DThR2_u'][i]=D[15]
        data2['DThp2_u'][i]=D[16]
        data2['DThz2_u'][i]=D[17]
        data2['Thpar2_u'][i]=D[20]
        data2['Thperp2_u'][i]=D[21]
    return data2
```
