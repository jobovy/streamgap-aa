```python
def load_denis_file(filename):
    names = ['x','y','z','vx','vy','vz','id']
    data = pd.read_csv(filename,names=names,comment='#',sep=' ')
    data['R']=np.sqrt(data['x']**2+data['y']**2)
    data['phi']=np.arctan2(data['y'],data['x'])
    data['vR']=(data['x']*data['vx']+data['y']*data['vy'])/data['R']
    data['vphi']=(-data['y']*data['vx']+data['x']*data['vy'])/data['R']
    data['E']=map(lambda i:pot.energy(i[:6]),data.values)
    data['Lz']=np.fabs(data['vphi']*data['R'])
    return data
```

```python
Orb = aa.Orbit(pot,1e-10)
```

```python
def integrate(x,t,frac=1):
    dT = frac*pot.torb(x)
    r = np.zeros((t/dT+2,6))
    r[0]=x
    i,T=1,0
    while(T<t):
        r[i]=Orb.integrate(r[i-1],dT,dT,False)
        T+=dT
        i+=1
    return r,T
```
