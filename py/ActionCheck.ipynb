```python
pot = aa.Logarithmic(220.,1.,0.9)
AA = aa.Actions_Genfunc(pot,"axisymmetric")
Orb = aa.Orbit(pot,1e-10)
```

```python
def load_denis_file(filename):
    names = ['x','y','z','vx','vy','vz','id']
    data = pd.read_csv(filename,names=names,comment='#',sep=' ')
    data['R']=np.sqrt(data['x']**2+data['y']**2)
    data['phi']=np.arctan2(data['y'],data['x'])
    data['vR']=(data['x']*data['vx']+data['y']*data['vy'])/data['R']
    data['vphi']=(-data['y']*data['vx']+data['x']*data['vy'])/data['R']
    data['E']=map(lambda i:pot.energy(i[:6]),data.values)
    data['Lz']=np.fabs(data['vphi']*data['R'])
    return data
```

```python
peri_1e8_data = load_denis_file('/data/jls/stream_gaps/tilted/snapshot_2nd_peri_1e8.txt')
point = peri_1e8_data.T[100]
```

```python
def integrate(x,t,frac=1,dt=None):
    if(dt==None):
        dT = frac*pot.torb(x)
    else:
        dT=dt
    r = np.zeros((t/dT+2,6))
    r[0]=x
    i,T=1,0
    while(T<t):
        r[i]=Orb.integrate(r[i-1],dT,dT,False)
        T+=dT
        i+=1
    return r,T
```

```python
R,T = integrate(point[:6].values,300.)
t = np.linspace(0.,T,len(R))
plt.plot(np.sqrt(R.T[0]**2+R.T[1]**2),R.T[2])
plt.xlabel('R');plt.ylabel('z')
```
