#### Load data

```python
def load_denis_file(filename):
    names = ['x','y','z','vx','vy','vz','id']
    data = pd.read_csv(filename,names=names,comment='#',sep=' ')
    data['R']=np.sqrt(data['x']**2+data['y']**2)
    data['phi']=np.arctan2(data['y'],data['x'])
    data['vR']=(data['x']*data['vx']+data['y']*data['vy'])/data['R']
    data['vphi']=(-data['y']*data['vx']+data['x']*data['vy'])/data['R']
    data['E']=map(lambda i:pot.energy(i[:6]),data.values)
    data['Lz']=np.fabs(data['vphi']*data['R'])
    return data
```

```python
peri_1e8_data = load_denis_file('/data/jls/stream_gaps/snapshot_2nd_peri_1e8.txt')
apo_1e8_data = load_denis_file('/data/jls/stream_gaps/snapshot_2nd_apo_1e8.txt')
peri_data = load_denis_file('/data/jls/stream_gaps/snapshot_2nd_peri_unperturbed.txt')
apo_data = load_denis_file('/data/jls/stream_gaps/snapshot_2nd_apo_unperturbed.txt')
```

```python
gap_centre_apo = np.array([-29.558912,-1.205861,1e-5,8.082564,-149.191223,1e-5])
gap_centre_peri = np.array([4.834022,14.219660,1e-5,-266.800964,129.468765,1e-5])
```

```python
def spatial_trim(data,centre,slop):
    phi = np.arctan2(centre[1],centre[0])
    data['dphi']=np.fmod(data['phi']-phi,2.*np.pi)
    data['dphi']-=2.*np.pi*(data['dphi']>np.pi)
    data = data[np.fabs(data.dphi)<slop]
    return data
```

```python
peri_1e8_data_trim = spatial_trim(peri_1e8_data,gap_centre_peri,np.deg2rad(30.))
apo_1e8_data_trim = spatial_trim(apo_1e8_data,gap_centre_apo,np.deg2rad(20.))
peri_data_trim = spatial_trim(peri_data,gap_centre_peri,np.deg2rad(30.))
apo_data_trim = spatial_trim(apo_data,gap_centre_apo,np.deg2rad(20.))
```

#### Find angles, freqs

```python
def add_actions_angles_freqs(data2,centre):
    data = data2
    data.is_copy=False    
    data['JR']=data['x']
    data['Jz']=data['x']
    data['angR']=data['x']
    data['angp']=data['x']
    data['angz']=data['x']
    data['OmR']=data['x']
    data['Omp']=data['x']
    data['Omz']=data['x']
    for i,d in zip(data.index,data.values):
        aa = AA.angles(d[:6])
        aa2 = AA.actions(d[:6])
        data['angR'][i],data['angp'][i],data['angz'][i]=aa[:3]
        data['OmR'][i],data['Omp'][i],data['Omz'][i]=aa[3:]
        data['JR'][i]=aa2[0];data['Jz'][i]=aa2[2]
    centre_aa = AA.angles(centre)
    centre_aa2 = AA.actions(centre)
    data['dangR']=data['angR']-centre_aa[0]
    data['dangp']=data['angp']-centre_aa[1]
    data['dangz']=data['angz']-centre_aa[2]
    data['dangR']-=2.*np.pi*(data['dangR']>np.pi)
    data['dangp']-=2.*np.pi*(data['dangp']>np.pi)
    data['dangz']-=2.*np.pi*(data['dangz']>np.pi)
    data['dOmR']=data['OmR']-centre_aa[3]
    data['dOmp']=data['Omp']-centre_aa[4]
    data['dOmz']=data['Omz']-centre_aa[5]
    data['dJR']=data['JR']-centre_aa2[0]
    data['dLz']=data['Lz']-np.fabs(centre_aa2[1])
    data['dJz']=data['Jz']-centre_aa2[2]    
    data['dE']=data['E']-pot.energy(centre)
    return data
```

```python
peri_1e8_data_trim_aa = add_actions_angles_freqs(peri_1e8_data_trim,gap_centre_peri)
apo_1e8_data_trim_aa = add_actions_angles_freqs(apo_1e8_data_trim,gap_centre_apo)
peri_data_trim_aa = add_actions_angles_freqs(peri_data_trim,gap_centre_peri)
apo_data_trim_aa = add_actions_angles_freqs(apo_data_trim,gap_centre_apo)
```

```python
def make_psplot(Data,title=None,mS=None,Alpha=None):
    ms,alpha = 0.5,0.3
    if(mS):
        ms=mS
    if(Alpha):
        alpha=Alpha
    f,a=plt.subplots(2,3,figsize=(12,8))
    plt.subplots_adjust(wspace=0.45,hspace=0.3)
    for data in Data:
        a[0][0].plot(data['x'],data['y'],'.',ms=ms,alpha=alpha)
        a[0][0].set_xlabel(labels['x']);a[0][0].set_ylabel(labels['y']);a[0][0].set_aspect('equal')
        a[0][1].plot(data['R'],data['z'],'.',ms=ms,alpha=alpha)
        a[0][1].set_xlabel(labels['R']);a[0][1].set_ylabel(labels['z'])
        a[0][2].plot(data['R'],data['vR'],'.',ms=ms,alpha=alpha)
        a[0][2].set_xlabel(labels['R']);a[0][2].set_ylabel(labels['vR'])
        a[1][0].plot(data['R'],data['vphi'],'.',ms=ms,alpha=alpha)
        a[1][0].set_xlabel(labels['R']);a[1][0].set_ylabel(labels['vphi'])
        a[1][1].plot(data['E']/10000.,data['Lz'],'.',ms=ms,alpha=alpha)
        a[1][1].set_xlabel(labels['E100']);a[1][1].set_ylabel(labels['Lz'])
    if title:
        plt.text(0.,1.02,title,horizontalalignment='left',verticalalignment='bottom',transform=a[0][0].transAxes,fontsize=16)
```
