```python
def load_denis_file(filename):
    names = ['x','y','z','vx','vy','vz','id']
    data = pd.read_csv(filename,names=names,comment='#',sep=' ')
    data['R']=np.sqrt(data['x']**2+data['y']**2)
    data['phi']=np.arctan2(data['y'],data['x'])
    data['vR']=(data['x']*data['vx']+data['y']*data['vy'])/data['R']
    data['vphi']=(-data['y']*data['vx']+data['x']*data['vy'])/data['R']
    data['E']=map(lambda i:pot.energy(i[:6]),data.values)
    data['Lz']=np.fabs(data['vphi']*data['R'])
    return data
```

```python
def spatial_trim(data,centre,slop):
    phi = np.arctan2(centre[1],centre[0])
    data['dphi']=np.fmod(data['phi']-phi,2.*np.pi)
    data['dphi']-=2.*np.pi*(data['dphi']>np.pi)
    data = data[np.fabs(data.dphi)<slop]
    return data
```

#### Find angles, freqs

```python
def add_actions_angles_freqs(data2,centre):
    data = data2
    data.is_copy=False    
    data['JR']=data['x']
    data['Jz']=data['x']
    data['angR']=data['x']
    data['angp']=data['x']
    data['angz']=data['x']
    data['OmR']=data['x']
    data['Omp']=data['x']
    data['Omz']=data['x']
    for i,d in zip(data.index,data.values):
        aa = AA.angles(d[:6])
        aa2 = AA.actions(d[:6])
        data['angR'][i],data['angp'][i],data['angz'][i]=aa[:3]
        data['OmR'][i],data['Omp'][i],data['Omz'][i]=aa[3:]
        data['JR'][i]=aa2[0];data['Jz'][i]=aa2[2]
    centre_aa = AA.angles(centre)
    centre_aa2 = AA.actions(centre)
    data['dangR']=data['angR']-centre_aa[0]
    data['dangp']=data['angp']-centre_aa[1]
    data['dangz']=data['angz']-centre_aa[2]
    data['dangR']-=2.*np.pi*(data['dangR']>np.pi)
    data['dangp']-=2.*np.pi*(data['dangp']>np.pi)
    data['dangz']-=2.*np.pi*(data['dangz']>np.pi)
    data['dOmR']=data['OmR']-centre_aa[3]
    data['dOmp']=data['Omp']-centre_aa[4]
    data['dOmz']=data['Omz']-centre_aa[5]
    data['dJR']=data['JR']-centre_aa2[0]
    data['dLz']=data['Lz']-np.fabs(centre_aa2[1])
    data['dJz']=data['Jz']-centre_aa2[2]    
    data['dE']=data['E']-pot.energy(centre)
    return data
```

```python
def make_psplot(Data,title=None,mS=None,Alpha=None):
    ms,alpha = 0.5,0.3
    if(mS):
        ms=mS
    if(Alpha):
        alpha=Alpha
    f,a=plt.subplots(2,3,figsize=(12,8))
    plt.subplots_adjust(wspace=0.45,hspace=0.3)
    for data in Data:
        a[0][0].plot(data['x'],data['y'],'.',ms=ms,alpha=alpha)
        a[0][0].set_xlabel(labels['x']);a[0][0].set_ylabel(labels['y']);a[0][0].set_aspect('equal')
        a[0][1].plot(data['R'],data['z'],'.',ms=ms,alpha=alpha)
        a[0][1].set_xlabel(labels['R']);a[0][1].set_ylabel(labels['z'])
        a[0][2].plot(data['R'],data['vR'],'.',ms=ms,alpha=alpha)
        a[0][2].set_xlabel(labels['R']);a[0][2].set_ylabel(labels['vR'])
        a[1][0].plot(data['R'],data['vphi'],'.',ms=ms,alpha=alpha)
        a[1][0].set_xlabel(labels['R']);a[1][0].set_ylabel(labels['vphi'])
        a[1][1].plot(data['E']/10000.,data['Lz'],'.',ms=ms,alpha=alpha)
        a[1][1].set_xlabel(labels['E100']);a[1][1].set_ylabel(labels['Lz'])
    if title:
        plt.text(0.,1.02,title,horizontalalignment='left',verticalalignment='bottom',transform=a[0][0].transAxes,fontsize=16)
```

```python
def rewind_snapshot(data2,t):
    data=data2
    data2['angR0']=np.fmod(data['angR']-t*data['OmR'],2.*np.pi)
    data2['angp0']=np.fmod(data['angp']-t*data['Omp'],2.*np.pi)
    data2['angz0']=np.fmod(data['angz']-t*data['Omz'],2.*np.pi)
    data2['dangR0']=data['dangR']-t*data['dOmR']
    data2['dangp0']=data['dangp']-t*data['dOmp']
    data2['dangz0']=data['dangz']-t*data['dOmz']
    return data
```

```python
from scipy.interpolate import UnivariateSpline, interp1d
from scipy.stats import linregress
class angle_freq_kicks:
            
    def dOmAngdv(self,x,dv):
        A=AA.angles(x)
        dOdv=np.zeros((3,3))
        dAdv=np.zeros((3,3))
        AOu=np.zeros((3,6))
        AOd=np.zeros((3,6))
        xu,xd=np.copy(x),np.copy(x)
        xu[3]+=dv
        xd[3]-=dv
        AOu[0]=AA.angles(xu)
        AOd[0]=AA.angles(xd)
        xu,xd=np.copy(x),np.copy(x)
        xu[4]+=dv
        xd[4]-=dv
        AOu[1]=AA.angles(xu)
        AOd[1]=AA.angles(xd)
        xu,xd=np.copy(x),np.copy(x)
        xu[5]+=dv
        xd[5]-=dv
        AOu[2]=AA.angles(xu)
        AOd[2]=AA.angles(xd)
        dAdv=AOu.T[:3].T-AOd.T[:3].T
        dOdv=AOu.T[3:6].T-AOd.T[3:6].T
        l,e = np.linalg.eigh(dAdv.T/2./dv)
        return A[:3],A[3:],dAdv.T/2./dv,dOdv.T/2./dv,e[-2]

    def __init__(self,merged_peri):
        
        ## Build spline of track
        f,a=plt.subplots(2,5,figsize=(16,7));plt.subplots_adjust(wspace=0.4,hspace=0.3)
        xs = np.linspace(np.min(merged_peri['x_y']),np.max(merged_peri['x_y']),100)
        
        s=UnivariateSpline(merged_peri['x_y'],merged_peri['y_y'],s=len(merged_peri)*10)
        hi=a[0][0].plot(np.sort(merged_peri['x_y']),merged_peri['y_y'][np.argsort(merged_peri['x_y'])],'.',ms=3)
        a[0][0].plot(xs,s(xs));a[0][0].set_xlabel(labels['x']);a[0][0].set_ylabel(labels['y']);
             
        sz=UnivariateSpline(merged_peri['x_y'],merged_peri['z_y'],s=len(merged_peri)*10)
        hi=a[0][1].plot(np.sort(merged_peri['x_y']),merged_peri['z_y'][np.argsort(merged_peri['x_y'])],'.',ms=3)
        a[0][1].plot(xs,sz(xs));a[0][1].set_xlabel(labels['x']);a[0][1].set_ylabel(labels['z']);
        
        s1=UnivariateSpline(merged_peri['x_y'],merged_peri['vx_y'],s=len(merged_peri)*100)
        hi=a[0][2].plot(np.sort(merged_peri['x_y']),merged_peri['vx_y'][np.argsort(merged_peri['x_y'])],'.',ms=3)
        a[0][2].plot(xs,s1(xs));a[0][2].set_xlabel(labels['x']);a[0][2].set_ylabel(labels['vx']);
        
        s2=UnivariateSpline(merged_peri['x_y'],merged_peri['vy_y'],s=len(merged_peri)*10)
        hi=a[0][3].plot(np.sort(merged_peri['x_y']),merged_peri['vy_y'][np.argsort(merged_peri['x_y'])],'.',ms=3)
        a[0][3].plot(xs,s2(xs));a[0][3].set_xlabel(labels['x']);a[0][3].set_ylabel(labels['vy']);
             
        s2z=UnivariateSpline(merged_peri['x_y'],merged_peri['vz_y'],s=len(merged_peri)*10)
        hi=a[0][4].plot(np.sort(merged_peri['x_y']),merged_peri['vz_y'][np.argsort(merged_peri['x_y'])],'.',ms=3)
        a[0][4].plot(xs,s2z(xs));a[0][4].set_xlabel(labels['x']);a[0][4].set_ylabel(labels['vz']);
        
        stream_track = np.ones((len(xs),6))*1e-5
        for i,j in enumerate(xs):
            stream_track[i][0]=j
            stream_track[i][1]=s(j)
            stream_track[i][2]=sz(j)
            stream_track[i][3]=s1(j)
            stream_track[i][4]=s2(j)
            stream_track[i][5]=s2z(j)
        a[0][0].plot(stream_track.T[0],stream_track.T[1])
        a[0][1].plot(stream_track.T[0],stream_track.T[2])
        a[0][2].plot(stream_track.T[0],stream_track.T[3])
        a[0][3].plot(stream_track.T[0],stream_track.T[4])
        a[0][4].plot(stream_track.T[0],stream_track.T[5])

        dOdAdv=map(lambda i:self.dOmAngdv(i,.6),stream_track)
        self.dOdAdv = dOdAdv
        b=0
        w=np.array([6.822,132.77,149.417])
        G=4.300918e-6
        M=1e8
        rs=0.625
        self.dv = galpy.df_src.streamgapdf.impulse_deltav_plummer_general(stream_track.T[3:].T,stream_track.T[:3].T,b,w,gap_centre_impact[:3],gap_centre_impact[3:],G*M,rs)
        self.dA = np.array([map(lambda i,j:np.dot(i[2],j),dOdAdv,self.dv)])[0]
        self.dO = np.array([map(lambda i,j:np.dot(i[3],j),dOdAdv,self.dv)])[0]
        self.O =  np.array([map(lambda i:i[1],dOdAdv)])[0]
        self.A = np.array([map(lambda i:i[0],dOdAdv)])[0]
        self.e = np.array([map(lambda i:i[4],dOdAdv)])[0]
        for i in range(3):
            self.A.T[i]=self.A.T[i]-2.*np.pi*(self.A.T[i]>np.pi)
        a[1][0].plot(self.A.T[0],self.A.T[1])
        a[1][1].plot(self.A.T[0],self.A.T[2])
        a[1][2].plot(self.O.T[0],self.O.T[1])
        a[1][3].plot(self.O.T[1],self.O.T[2])
        slope, intercept, r_value, p_value, std_err = linregress(self.O.T[0],self.O.T[1])
        slope2, intercept2, r_value, p_value, std_err = linregress(self.O.T[0],self.O.T[2])
        self.n = np.array([1.,slope,slope2])
        a[1][2].plot(self.O.T[0],self.n[1]*self.O.T[0]+intercept)
        a[1][3].plot(self.n[1]*self.O.T[0]+intercept,self.n[2]*self.O.T[0]+intercept2)
        self.n = self.n/np.sqrt(self.n)
        self.angs_centre = AA.angles(gap_centre_impact)
        self.angs_centre[2]-=2.*np.pi
        self.deltaang_par=np.dot(self.A-self.angs_centre[:3],self.n)
        self.deltaang_par=self.deltaang_par-2.*np.pi*(self.deltaang_par>np.pi)
        
        a[1][4].plot(self.A.T[0])
        self.angR_k=interp1d(self.deltaang_par,self.dA.T[0])
        self.angp_k=interp1d(self.deltaang_par,self.dA.T[1])
        self.angz_k=interp1d(self.deltaang_par,self.dA.T[2])
        self.OmR_k=interp1d(self.deltaang_par,self.dO.T[0])
        self.Omp_k=interp1d(self.deltaang_par,self.dO.T[1])
        self.Omz_k=interp1d(self.deltaang_par,self.dO.T[2])
    
    def kick(self,ang):
        return self.angR_k(ang),self.angp_k(ang),self.angz_k(ang),self.OmR_k(ang),self.Omp_k(ang),self.Omz_k(ang)
```

```python
def fastforward_snapshot(data2,t):
    data=data2
    data2['angR_now']=np.fmod(data['angR_y']+data['angR_k']+t*(data['OmR_y']+data['OmR_k']),2.*np.pi)
    data2['angp_now']=np.fmod(data['angp_y']+data['angp_k']+t*(data['Omp_y']+data['Omp_k']),2.*np.pi)
    data2['angz_now']=np.fmod(data['angz_y']+data['angz_k']+t*(data['Omz_y']+data['Omz_k']),2.*np.pi)
    data2['OmR_now']=(data['OmR_y']+data['OmR_k'])
    data2['Omp_now']=(data['Omp_y']+data['Omp_k'])
    data2['Omz_now']=(data['Omz_y']+data['Omz_k'])
    data2['angR_now_unp']=np.fmod(data['angR_y']+t*(data['OmR_y']),2.*np.pi)
    data2['angp_now_unp']=np.fmod(data['angp_y']+t*(data['Omp_y']),2.*np.pi)
    data2['angz_now_unp']=np.fmod(data['angz_y']+t*(data['Omz_y']),2.*np.pi)
    return data
```
